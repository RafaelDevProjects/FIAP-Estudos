# **FUNÇÕES**

```python
from sympy import *
```

```python
from sympy.plotting import *
```

```python
x = Symbol('x')
```

# **FUNÇÕES TRIGONOMÉTRICAS**

```python
plot(sin(x));
```

```python
plot(3*sin(x)); # o 3 é a amplitude // o x é o argumento
```

```python
plot(cos(x));
```

```python
plot(2*cos(x));
```

```python
plot(tan(x), (x,0,10));
```

```python
plot(100-20*cos((8*3.14/3)*x), (x,0,2));
```

```python
plot(sin(x+pi)) # O +pi é oque inddica defazagem ou avanço de fase
```
